```python
### NULL HYPOTHESIS
# There is no significant association between category bins and customer purchasing behavior (one-time or repeat customer).

### ALTERNATIVE HYPOTHEIS 
# There is a significant association between category bins and customer purchasing behavior.

from scipy.stats import chi2_contingency

chi_2,p,dof,expected = chi2_contingency(contigency_table)

if p<0.05:
    print('We Reject Null Hypothesis That Leads To The Conclusion That There Is A Strong Relationship Between Customer Type & Category Bins')
else:
    print('We Fail To Reject Null Hypothesis There Is Very Low Realtion Between Customer_Type & Category Bin')
```

# HYPOTHESIS TESTING 

```python
## NULL HYPOTHESIS 
# There is no significant difference between Customer_type(repeat_cust,one_time_cust) in case of there intial basket size.

## ALTERNATIVE HYPOTHESIS 
# These is significant difference between Customer_type(repeat_cust,one_time_cust) in case of there intial basket size.

from scipy.stats import ttest_ind

t_stat,p = ttest_ind(Repeat_cust['Total_order_items'],One_Time_cust['Total_order_items'])

if p<0.05:
    print('We Reject Null Hypothesis there is a significant difference between intial basket size of repeat & one time customer')

else :
    print('We Accept Null Hypothesis there is a no significant difference between intial basket size of repeat & one time customer')
```
